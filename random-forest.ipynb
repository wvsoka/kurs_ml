# Random forest intuition

Step 1: pick at random k data points from the training dataset
step 2: build DT associated to these k data points
step 3: choose the number ntree of trees you want to bild and repeat steps 1 & 2
step 4: for a new data point make each one of your ntree trees predict the value of Y to for the data point in questuin and assing the new data point the average across all the predicted Y values

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Machine-Learning-A-Z-Codes-Datasets/Part 2 - Regression/Section 9 - Random Forest Regression/Python/Position_Salaries.csv')

X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=10, random_state=0)
model.fit(X, y)
```

```python
model.predict([[6.5]])
```

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color='red')
plt.plot(X_grid, model.predict(X_grid), color='blue')
plt.show()
```
