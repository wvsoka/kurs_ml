# evaluating regression models performance

## R squared:
- residual sum of squares SS_res = SUM(y_i - y^_i)^2
- total sum of squares SS_tot = SUM(y_i - y_avg)^2

R^2 = 1 - SS_res/SS_tot

we want to minimalize SS_res. usually ss_res<ss_tot (not always).

rule of thumb:
- r2 = 1 <- perfect fit
- ~0.9 very good
- <0.7 not great
- <0.4 terrible
- 0 model makes no sense for this data
